```r
j = seq(-2, 2, 0.1)
k = (1/2)*cos(2*pi*j)+(7/2)
s <- seq(1, 6, 0.1)
t <- 12/(s+1)
u <- seq(-6, -1, 0.1)
v <- 12/(-u+1)
x <- seq(-1, 1, 0.1)
y <- x^4-x^2+6
plot(j, k, ylim=c(0, 8), xlim=c(-8, 8), type="l", xlab="", ylab="")
par(new=T)
plot(s, t, ylim=c(0, 8), xlim=c(-8, 8), type="l", xlab="", ylab="")
par(new=T)
plot(u, v, ylim=c(0, 8), xlim=c(-8, 8), type="l", xlab="", ylab="")
par(new=T)
plot(x, y, ylim=c(0, 8), xlim=c(-8, 8), type="l", xlab="", ylab="")
```
